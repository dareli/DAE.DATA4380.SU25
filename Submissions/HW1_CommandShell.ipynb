### Cancer Patient Dataset

- **Sort the cancer patient by age**

head cancer_patient_dataset.csv

(head -n 1 "cancer_patient_dataset.csv" && tail -n +2 "cancer_patient_dataset.csv" | sort -t',' -k3 -n) > cancer_sorted_by_age.csv

head cancer_sorted_by_age.csv

- notes: 
  - outer parentheses () group the header and sorted data
  - head -n 1: extract header | tail -n +2: skips header & shows data rows only
  - -k3 = sort 3rd col, n = numeric sort
