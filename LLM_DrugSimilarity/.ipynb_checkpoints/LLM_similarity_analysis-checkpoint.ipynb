# LLM Similarity Analysis

This notebook compares drug description similarity (using cosine) to side effect overlap (using Jaccard). Tested 1000 random drug pairs and found that most pairs shared no side effects, even if their descriptions were similar.

Findings:
- Pearson correlation (all pairs): **-0.024**
- Pearson correlation (filtered 3+ side effects): **-0.077**
- Clustering the embeddings formed 7 distinct groups (visualized with PCA)

------

```python
# imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import jaccard
import random
from scipy.stats import spearmanr, pearsonr
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# load prepared data

df = pd.read_csv("drug_descriptions_embeddings.csv")

# recover  the embedding vectors
embedding_cols = [col for col in df.columns if col.startswith("emb_")]
df['embedding'] = df[embedding_cols].values.tolist()

# take side_effect_name col from string to set
import ast
df['side_effect_name'] = df['side_effect_name'].apply(lambda x: set(ast.literal_eval(x)))

print("Data loaded:", df.shape)
print("Example drug:")
print(df[['name', 'side_effect_name']].iloc[0])
```

```python
# compute cosine and jaccard similarity

N = 1000
random.seed(42)
pairs = []

for _ in range(N):
    i, j = random.sample(range(len(df)), 2)
    emb_i = df.iloc[i]['embedding']
    emb_j = df.iloc[j]['embedding']
    se_i = df.iloc[i]['side_effect_name']
    se_j = df.iloc[j]['side_effect_name']

    cos_sim = cosine_similarity([emb_i], [emb_j])[0][0]
    try:
        jaccard_sim = 1 - jaccard(list(se_i), list(se_j)) if se_i or se_j else 0
    except:
        jaccard_sim = 0

    pairs.append((cos_sim, jaccard_sim))
```

```python
cos_sims, jaccard_sims = zip(*pairs)

plt.figure(figsize=(8, 6))
plt.scatter(cos_sims, jaccard_sims, alpha=0.4)
plt.xlabel("Cosine Similarity (Descriptions)")
plt.ylabel("Jaccard Similarity (Side Effects)")
plt.title("Cosine vs. Jaccard Similarity for 1000 Drug Pairs")
plt.grid(True)
plt.tight_layout()
plt.savefig("simana_cosjac.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# look at correlation
pearson_corr, _ = pearsonr(cos_sims, jaccard_sims)
spearman_corr, _ = spearmanr(cos_sims, jaccard_sims)

print(f"Pearson correlation: {pearson_corr:.3f}")
print(f"Spearman correlation: {spearman_corr:.3f}")
```

### notes: 
- most drug pairs shared no side effects, even if their descriptions were similar.
- the correlation between the two isvery weak (Pearson: -0.024, Spearman: -0.034). This means description similarity doesn't really predict side effect similarity.


### doing some resampling

```python
# this time i will filter for drugs with 3+ side effects
filtered_df = df[df['side_effect_name'].apply(lambda x: len(x) >= 3)].reset_index(drop=True)
print(f"Filtered to {len(filtered_df)} drugs with AT LEAST 3 side effects")
```

```python
# resample and compute similarities

N = 1000
random.seed(42)
resampled_pairs = []

for _ in range(N):
    i, j = random.sample(range(len(filtered_df)), 2)
    emb_i = filtered_df.iloc[i]['embedding']
    emb_j = filtered_df.iloc[j]['embedding']
    se_i = filtered_df.iloc[i]['side_effect_name']
    se_j = filtered_df.iloc[j]['side_effect_name']

    cos_sim = cosine_similarity([emb_i], [emb_j])[0][0]
    try:
        jaccard_sim = 1 - jaccard(list(se_i), list(se_j)) if se_i or se_j else 0
    except:
        jaccard_sim = 0

    resampled_pairs.append((cos_sim, jaccard_sim))
```

```python
cos_sims_r, jaccard_sims_r = zip(*resampled_pairs)

plt.figure(figsize=(8, 6))
plt.scatter(cos_sims_r, jaccard_sims_r, alpha=0.4)
plt.xlabel("Cosine Similarity (Descriptions)")
plt.ylabel("Jaccard Similarity (Side Effects)")
plt.title("Filtered: Cosine vs. Jaccard Similarity (3 SEs)")
plt.grid(True)
plt.tight_layout()
plt.savefig("resample_cosjac.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
pearson_r, _ = pearsonr(cos_sims_r, jaccard_sims_r)
spearman_r, _ = spearmanr(cos_sims_r, jaccard_sims_r)

print(f"Filtered Pearson correlation: {pearson_r:.3f}")
print(f"Filtered Spearman correlation: {spearman_r:.3f}")
```

#### notes:
- even among drugs with at least 3 side effects, there's still no meaningful correlation between how similar their descriptions are and how similar their side effects are
- result actually got slightly more negative

-----

### Trying Clustering now

```python
# extract and cluster the embeddings
X = np.vstack(df['embedding'].values) # get the raw embeddings as a matrix

k = 7 # number of clusters 
kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
df['cluster'] = kmeans.fit_predict(X)

print("Cluster counts:")
print(df['cluster'].value_counts())
```

```python
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

# plot em clusters
plt.figure(figsize=(8, 6))
for i in range(k):
    pts = X_2d[df['cluster'] == i]
    plt.scatter(pts[:, 0], pts[:, 1], label=f'Cluster {i}', alpha=0.6)

plt.title("Drug Description Clusters")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("clusters_1.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# ouuu now let me look into the clusters
for i in range(7):
    print(f"Cluster {i}:")
    print(df[df['cluster'] == i]['name'].sample(5).tolist())
    print()
```
