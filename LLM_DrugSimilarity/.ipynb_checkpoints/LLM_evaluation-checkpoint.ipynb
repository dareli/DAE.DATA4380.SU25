# LLM Drug Similarity â€“ Evaluation Notebook

This notebook checks how well drug description embeddings group similar drugs and match side effects. Then I compared the text similarity (cosine) with side effect overlap (Jaccard) and look at cluster patterns.

**Key results:**
- Clusters are weakly separated (Silhouette: 0.058)
- Cosine similarity is a bit higher within clusters than between them
- Side effect overlap is slightly higher within clusters, but still low
- Common effects like "Nausea" appear in all clusters

-------

```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import silhouette_score
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import adjusted_rand_score
from sklearn.decomposition import PCA
import random
from collections import Counter
import ast
```

```python
# load data and setup
# load clustered drug data w/ embeddings and side effects
clustered_df = pd.read_csv("drug_data_with_clusters.csv")

# load similarity scores
sim_all = pd.read_csv("similarity_scores_all.csv")
sim_filtered = pd.read_csv("similarity_scores_filtered.csv")

print("Loaded:")
print("\n\nClustered data:", clustered_df.shape)
print("\n\nAll similarity scores:", sim_all.shape)
print("\n\nFiltered similarity scores:", sim_filtered.shape)
```

```python
# re-plot similarity results
plt.figure(figsize=(6, 5))
plt.scatter(sim_all['cosine'], sim_all['jaccard'], alpha=0.4)
plt.xlabel("Cosine Similarity")
plt.ylabel("Jaccard Similarity")
plt.title("All Drug Pairs")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(sim_filtered['cosine'], sim_filtered['jaccard'], alpha=0.4)
plt.xlabel("Cosine Similarity")
plt.ylabel("Jaccard Similarity")
plt.title("Filtered Pairs (â‰¥3 side effects)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# cluster quality evaluation
embedding_cols = [col for col in clustered_df.columns if col.startswith("emb_")]
X = clustered_df[embedding_cols].values
labels = clustered_df['cluster']

# silhouette score
sil_score = silhouette_score(X, labels, metric='cosine')
print(f"Silhouette Score (cosine): {sil_score:.3f}")
```

```python
# again re-plot clusters 
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X)

plt.figure(figsize=(7, 6))
for i in sorted(clustered_df['cluster'].unique()):
    pts = X_2d[clustered_df['cluster'] == i]
    plt.scatter(pts[:, 0], pts[:, 1], label=f"Cluster {i}", alpha=0.6)

plt.title("Drug Embedding Clusters (PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

-------
#### At least look at top side effects per cluster now

```python
clustered_df['side_effect_name'] = clustered_df['side_effect_name'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
all_counts = Counter()
for se_set in clustered_df['side_effect_name']:
    all_counts.update(se_set)
top_effects = [se for se, _ in all_counts.most_common(20)] # top 20 most frequent side effects
clusters = sorted(clustered_df['cluster'].unique())
mat = np.zeros((len(top_effects), len(clusters)), dtype=int)
for ci, c in enumerate(clusters):
    subset = clustered_df[clustered_df['cluster'] == c]['side_effect_name']
    c_counts = Counter()
    for s in subset:
        c_counts.update(s)
    for ri, se in enumerate(top_effects):
        mat[ri, ci] = c_counts.get(se, 0)
```

```python
fig, ax = plt.subplots(figsize=(10, max(6, 0.4*len(top_effects))))
im = ax.imshow(mat, aspect='auto')
ax.set_xticks(range(len(clusters)))
ax.set_xticklabels([f"C{c}" for c in clusters])
ax.set_yticks(range(len(top_effects)))
ax.set_yticklabels(top_effects)
ax.set_xlabel("Cluster")
ax.set_ylabel("Side effect")
ax.set_title("Top side effects by cluster (counts)")
fig.colorbar(im, ax=ax, label="Count")
plt.savefig("clusters_topse.png", dpi=300, bbox_inches='tight')
plt.tight_layout(); plt.show()
```

```python
# within vs between-cluster jaccard similarity

# to compute Jaccard similarity for sets
def jaccard_sets(a, b):
    if not a and not b:
        return 0.0
    inter = len(a & b)
    union = len(a | b)
    return inter / union if union else 0.0

# indices by cluster all grouped
idx_by_cluster = {c: clustered_df.index[clustered_df['cluster'] == c].tolist() for c in clusters}
all_indices = clustered_df.index.tolist()

same_scores, diff_scores = [], []

for c, idxs in idx_by_cluster.items(): # sample within ea cluster for scores
    if len(idxs) < 2:
        continue
    for _ in range(min(150, len(idxs)*(len(idxs)-1)//2)):
        i, j = random.sample(idxs, 2)
        same_scores.append(jaccard_sets(clustered_df.at[i, 'side_effect_name'], clustered_df.at[j, 'side_effect_name']))

count = 0
while count < 1000:
    i, j = random.sample(all_indices, 2)
    if clustered_df.at[i, 'cluster'] != clustered_df.at[j, 'cluster']:
        diff_scores.append(jaccard_sets(clustered_df.at[i, 'side_effect_name'], clustered_df.at[j, 'side_effect_name']))
        count += 1
```

```python
# plot comparison of within vs betweencluster similarities
plt.figure(figsize=(7, 5))
plt.boxplot([same_scores, diff_scores], labels=["same cluster", "different clusters"], showfliers=False)
plt.ylabel("jaccard similarity (side effects)")
plt.title("within vs between cluster sideeffect similarity")
plt.grid(True, axis='y')
plt.savefig("jac_with_se.png", dpi=300, bbox_inches='tight')
plt.tight_layout(); plt.show()

print(f"mean jaccard (same cluster): {np.mean(same_scores):.3f} (n={len(same_scores)})")
print(f"mean jaccard (diff clusters): {np.mean(diff_scores):.3f} (n={len(diff_scores)})")
```

-------
#### extra insights and interpertability

```python
# trace a specific side effect through clusters

target_effect = "Nausea"  # can change to any side effect, this is just an example
has_effect = clustered_df[clustered_df['side_effect_name'].apply(lambda se: target_effect in se)]
print(f"'{target_effect}' found in {len(has_effect)} drugs across clusters:")
print(has_effect['cluster'].value_counts().sort_index())


for cluster_id in sorted(has_effect['cluster'].unique()):
    print(f"\nCluster {cluster_id} examples:")
    print(has_effect[has_effect['cluster'] == cluster_id]['name'].head(5).tolist())
```

```python
# top 5 side effects per cluster
for cluster_id in sorted(clustered_df['cluster'].unique()):
    effects_counter = Counter()
    clustered_df[clustered_df['cluster'] == cluster_id]['side_effect_name'].apply(effects_counter.update)
    top_effects = [f"{effect} ({count})" for effect, count in effects_counter.most_common(5)]
    print(f"Cluster {cluster_id}: {', '.join(top_effects)}")
```

-------
#### extra metrics

```python
# purity of top side effect in each cluster
for cluster_id in sorted(clustered_df['cluster'].unique()):
    effects_counter = Counter()
    subset = clustered_df[clustered_df['cluster'] == cluster_id]
    subset['side_effect_name'].apply(effects_counter.update)
    top_effect, count = effects_counter.most_common(1)[0]
    purity = count / len(subset)
    print(f"Cluster {cluster_id}: Top SE = {top_effect} | Purity = {purity:.2%}")
```

```python
# example binary label... has target_effect or not
labels_true = clustered_df['side_effect_name'].apply(lambda se: target_effect in se).astype(int)
labels_pred = clustered_df['cluster']

ari = adjusted_rand_score(labels_true, labels_pred)
print(f"ARI for '{target_effect}': {ari:.3f}")
```

```python
# cosine similarity matrix
cos_sim_matrix = cosine_similarity(X)

within_sims, between_sims = [], []
for i in range(len(clustered_df)):
    for j in range(i+1, len(clustered_df)):
        if clustered_df.at[i, 'cluster'] == clustered_df.at[j, 'cluster']:
            within_sims.append(cos_sim_matrix[i, j])
        else:
            between_sims.append(cos_sim_matrix[i, j])

print(f"mean within cluster cosine similarity: {np.mean(within_sims):.3f}")
print(f"mean between cluster cosine similarity: {np.mean(between_sims):.3f}")
```

```python
# summary table

metrics_summary = {}

# silhouette score
metrics_summary['Silhouette (cosine)'] = sil_score

# mean jaccard similarities
metrics_summary['Mean Jaccard (same cluster)'] = np.mean(same_scores)
metrics_summary['Mean Jaccard (diff cluster)'] = np.mean(diff_scores)

# within vs between cluster cos similarity
metrics_summary['Mean Cosine (same cluster)'] = np.mean(within_sims)
metrics_summary['Mean Cosine (diff cluster)'] = np.mean(between_sims)

# adjusted rand index for specific side effect
target_effect = "Nausea"
labels_true = clustered_df['side_effect_name'].apply(lambda se: target_effect in se).astype(int)
labels_pred = clustered_df['cluster']
metrics_summary[f"ARI ('{target_effect}')"] = adjusted_rand_score(labels_true, labels_pred)

metrics_df = pd.DataFrame(metrics_summary, index=['Value']).T
metrics_df.columns = ['Value']
print(metrics_df)
```
